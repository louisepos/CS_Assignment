```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

%matplotlib notebook
```

```python
data = pd.read_csv("istherecorrelation.csv", delimiter=";", decimal=",")
```

```python
data
```

```python
X = data.iloc[:,0]   #Years
Y = data.iloc[:,1]   #WO
Z = data.iloc[:,2]   #NL Beer Consumption
```

```python
fig, ax = plt.subplots()
fig.set_dpi(300)
ax.plot(X, Y, label='line1', color='red', marker='o')
ax.set_xlabel("Year")
ax.set_ylabel("WO [x1000]", color='red')
```

```python
ax2 = ax.twinx()
ax2.plot(X, Z, color='blue', marker='o')
ax2.set_ylabel("NL Beer Consumption [x1000 hectoL]", color='blue')

plt.show()
```

```python
plt.savefig('solution_plot.png', bbox_inches='tight')
```
